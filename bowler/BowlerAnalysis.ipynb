```python
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
from pathlib import Path
filepath_batter_for_clustering = Path('bowlers_for_clustering.csv') 
bowler_train.to_csv(filepath_batter_for_clustering,index=False)

filepath_batter_for_clustering = Path('bowlers_test.csv') 
bowler_test.to_csv(filepath_batter_for_clustering,index=False)
```

```python
bowler_for_clustering = bowler.copy().drop(['name','team','prelim_metric','pre-rank'],axis=1)
columns = bowler_for_clustering.columns.tolist()
```

```python
wcss = []
for i in range(1,11,1):
  model = KMeans(i).fit(bowler_for_clustering)
  wcss.append(model.inertia_)
plt.plot(np.arange(1,11,1),wcss)
plt.show()
```

```python
plt.figure(figsize=(17,8)) 
sns.boxplot(data=bowler_for_clustering)
plt.show()
```

```python
fig = plt.figure(figsize = (15,10))
ax = fig.gca()
bowler_for_clustering.hist(ax=ax)
plt.show()
```

```python
range_n_clusters = [2, 3, 4, 5, 6,7,8,9,10]

for n_clusters in range_n_clusters:

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(bowler_for_clustering)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(bowler_for_clustering, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )
```

```python
corr_normal = bowler_for_clustering.corr()

# Setting up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Creating the heatmap using Matplotlib
cax = ax.matshow(corr_normal, cmap='coolwarm', vmin=-1, vmax=1)

# Adding color bar
fig.colorbar(cax)

# Setting ticks and labels
ax.set_xticks(np.arange(len(corr_normal.columns)))
ax.set_yticks(np.arange(len(corr_normal.columns)))
ax.set_xticklabels(corr_normal.columns, rotation=45, ha='left')
ax.set_yticklabels(corr_normal.columns)

# Annotating the heatmap with correlation values
for i in range(len(corr_normal.columns)):
    for j in range(len(corr_normal.columns)):
        text = ax.text(j, i, round(corr_normal.iloc[i, j], 2),
                       ha='center', va='center', color='black')

# Adding title to the heatmap
plt.title('Correlation Heatmap')

# Displaying the heatmap
plt.show()
```
