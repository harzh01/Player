```python
import pandas as pd
import numpy as np
```

```python
allrounder_name=pd.read_excel('All_Rounders.xlsx')
```

```python
names=allrounder_name['Player']
```

```python
names.dropna()
```

```python
bowler = pd.read_csv('bowling_records.csv')
batter=pd.read_csv('batting_records.csv')
```

```python
def get_player_records(bowler, batter, names):    
       
    bowler_records = bowler[bowler['name'].isin(names)]
    batter_records = batter[batter['name'].isin(names)]
    
    return bowler_records, batter_records
```

```python
bowler_records, batter_records = get_player_records(bowler, batter, names)
```

```python
from pathlib import Path
filepath_bowling = Path('all_bowling.csv') 
filepath_batting = Path('all_batting.csv') 

bowler_records.to_csv(filepath_bowling,index=False)
batter_records.to_csv(filepath_batting,index=False)
```

```python
bowler_records = bowler_records.drop_duplicates(subset=['name'])
```
